```python
import math
from collections import defaultdict

# Define the SKILL_ALIASES dictionary
SKILL_ALIASES = {
    "python": "python",
    "pyhton": "python",
    "java": "java",
    "javascript": "javascript",
    "javascrpit": "javascript",
    "js": "javascript",
    "typescript": "typescript",
    "typescrpit": "typescript",
    "c++": "cpp",
    "cpp": "cpp",
    "r": "r",
    "kotlin": "kotlin",
    "machinelearning": "machine_learning",
    "machine learning": "machine_learning",
    "ml": "machine_learning",
    "sklearn": "machine_learning",
    "deeplearning": "deep_learning",
    "deep learning": "deep_learning",
    "deep-learning": "deep_learning",
    "tensorflow": "tensorflow",
    "pytorch": "pytorch",
    "keras": "keras",
    "nlp": "nlp",
    "bert": "bert",
    "xgboost": "xgboost",
    "feature engineering": "feature_engineering",
    "statistics": "statistics",
    "stats": "statistics",
    "regression": "regression",
    "clustering": "clustering",
    "data-viz": "data_visualization",
    "data visualization": "data_visualization",
    "data viz": "data_visualization",
    "matplotlib": "data_visualization",
    "tableau": "data_visualization",
    "power-bi": "data_visualization",
    "power bi": "data_visualization",
    "powerbi": "data_visualization",
    "pandas": "pandas",
    "numpy": "numpy",
    "react": "react",
    "reacts": "react",
    "reactjs": "react",
    "vue": "vue",
    "vue.js": "vue",
    "vuejs": "vue",
    "redux": "redux",
    "tailwind": "tailwind",
    "html/css": "html_css",
    "html css": "html_css",
    "html": "html_css",
    "css": "html_css",
    "jest": "jest",
    "graphql": "graphql",
    "node.js": "nodejs",
    "nodejs": "nodejs",
    "node js": "nodejs",
    "flask": "flask",
    "spring boot": "spring_boot",
    "springboot": "spring_boot",
    "rest api": "rest_api",
    "rest": "rest_api",
    "restapi": "rest_api",
    "microservices": "microservices",
    "sql": "sql",
    "mysql": "mysql",
    "mysq": "mysql",
    "postgresql": "postgresql",
    "postgres": "postgresql",
    "mongodb": "mongodb",
    "redis": "redis",
    "docker": "docker",
    "kubernetes": "kubernetes",
    "kubernates": "kubernetes",
    "k8s": "kubernetes",
    "ci/cd": "ci_cd",
    "cicd": "ci_cd",
    "ci cd": "ci_cd",
    "aws": "aws",
    "android": "android",
    "firebase": "firebase",
    "algorithms": "algorithms",
    "algoritms": "algorithms",
    "data structure": "data_structures",
    "data structures": "data_structures",
    "competitive programming": "competitive_programming",
    "ui/ux": "ui_ux",
    "ui ux": "ui_ux",
    "figma": "figma"
}

# Define the resumes and job descriptions
resumes = [
    {"id": "01", "raw_skills": "Pyhton, MachineLearning, SQL, pandas, numpy, Deep-learning", "background": "TCS Intern · BITS Pilani CSE 2024"},
    {"id": "02", "raw_skills": "JavaScrpit, Reacts, Node.JS, MongoDb, REST api, HTML/CSS", "background": "Freelance Web Developer · VIT IT 2024"},
    {"id": "03", "raw_skills": "Java, Spring Boot, MySql, Microservices, Docker, kubernates", "background": "Infosys SDE Intern · IIT Delhi 2023"},
    {"id": "04", "raw_skills": "Python, TensorFlow, Keras, NLP, BERT, data-viz, matplotlib", "background": "IISc Research Assistant · IIIT Hyderabad AI 2024"},
    {"id": "05", "raw_skills": "C++, Algoritms, Data Structure, competitive programming, python", "background": "Google SWE Intern · IIT Bombay 2024"},
    {"id": "06", "raw_skills": "javascript, vue.js, python, flask, PostgreSQL, AWS, CI/CD", "background": "Full Stack Developer · NIT Trichy 2022"},
    {"id": "07", "raw_skills": "Python, Sklearn, XGboost, feature engineering, SQL, tableau", "background": "Data Analyst · Delhi University 2023"},
    {"id": "08", "raw_skills": "Java, Android, Kotlin, Firebase, REST, UI/UX, figma", "background": "Samsung Android Intern · NSIT 2024"},
    {"id": "09", "raw_skills": "Reactjs, TypeScrpit, GraphQL, redux, tailwind, nodejs, jest", "background": "Frontend SDE · Flipkart / IIIT Bangalore"},
    {"id": "10", "raw_skills": "python, R, statistics, ML, regression, clustering, Power-BI", "background": "Data Science Intern · Wipro 2024"},
]

job_descriptions = [
    {"id": "JD-1", "company": "Kakao (Seoul)", "role": "ML Engineer", "requiredskills": "Python, Machine Learning, Deep Learning, TensorFlow, PyTorch, SQL, Data Visualization", "preferredskills": "NLP, BERT, Feature Engineering, Statistics"},
    {"id": "JD-2", "company": "Naver (Seongnam)", "role": "Backend Engineer", "requiredskills": "Java, Spring Boot, MySQL, PostgreSQL, Microservices, Docker, Kubernetes", "preferredskills": "REST API, CI/CD, Redis"},
    {"id": "JD-3", "company": "Line (Seoul)", "role": "Frontend Engineer", "requiredskills": "JavaScript, React, Vue, TypeScript, REST API, HTML/CSS", "preferredskills": "Node.js, GraphQL, Redux, Jest, AWS"},
]

# Normalize the raw skills in each resume
for resume in resumes:
    raw_skills_str = resume["raw_skills"].lower().replace(",", " ").split() # Corrected key and variable name
    normalized_skills = []
    for skill in raw_skills_str: # Corrected variable name
        if skill in SKILL_ALIASES:
            normalized_skills.append(SKILL_ALIASES[skill]) # Corrected dictionary name
        else:
            normalized_skills.append(skill)
    resume["normalized_skills"] = list(set(normalized_skills)) # Corrected key and standardized variable name

# Normalize the required skills in each job description
for jobdescription in job_descriptions: # Standardized variable name
    required_skills_str = jobdescription["requiredskills"].lower().replace(",", " ").split() # Corrected key and variable name
    normalized_required_skills = [] # Standardized variable name
    for skill in required_skills_str: # Corrected variable name
        if skill in SKILL_ALIASES:
            normalized_required_skills.append(SKILL_ALIASES[skill])
        else:
            normalized_required_skills.append(skill)
    jobdescription["normalized_required_skills"] = list(set(normalized_required_skills)) # Corrected key and standardized variable name

# Compute the term frequency (TF) of each skill in each resume
for resume in resumes:
    tf = {}
    for skill in resume["normalized_skills"]:
        if skill not in tf:
            tf[skill] = 1
        else:
            tf[skill] += 1
    total_skills_count = len(resume["normalized_skills"]) # Standardized variable name
    for skill in tf:
        if total_skills_count > 0: # Avoid division by zero
            tf[skill] /= total_skills_count
        else:
            tf[skill] = 0 # Handle case where there are no skills
    resume["tf"] = tf

# Compute the inverse document frequency (IDF) of each skill
idfs = {}
all_skills = list(set(skill for resume in resumes for skill in resume["normalized_skills"])) # Collect all unique skills first
num_resumes = len(resumes)

for skill in all_skills:
    num_resumes_with_skill = sum(1 for resume in resumes if skill in resume["normalized_skills"]) # Standardized variable name
    if num_resumes_with_skill > 0:
        idf = math.log(num_resumes / num_resumes_with_skill)
    else:
        idf = 0 # Handle case where skill is not in any resume (should ideally not happen with 'all_skills')
    idfs[skill] = idf

# Compute the TF-IDF of each skill in each resume
for resume in resumes:
    tf_idf = {} # Standardized variable name
    for skill in resume["normalized_skills"]:
        tf_idf[skill] = resume["tf"][skill] * idfs.get(skill, 0) # Use .get with default 0 in case skill is not in idfs
    resume["tf_idf"] = tf_idf # Standardized key

# Compute the cosine similarity between each resume and job description
similarities = {}
for jobdescription in job_descriptions: # Standardized variable name
    job_similarities = []
    for resume in resumes:
        dot_product = 0
        magnitude_resume = 0
        magnitude_job = 0

        # Use the union of skills from resume's tf_idf and job description's normalized_required_skills
        combined_skills = set(resume["tf_idf"].keys()) | set(jobdescription["normalized_required_skills"])

        for skill in combined_skills:
            resume_tfidf_value = resume["tf_idf"].get(skill, 0)
            job_skill_present = 1 if skill in jobdescription["normalized_required_skills"] else 0

            dot_product += resume_tfidf_value * job_skill_present
            magnitude_resume += resume_tfidf_value ** 2 # Corrected square operation
            magnitude_job += job_skill_present ** 2 # Magnitude for job description (binary presence)

        # Calculate cosine similarity
        if magnitude_resume > 0 and magnitude_job > 0:
            cosine_similarity = dot_product / (math.sqrt(magnitude_resume) * math.sqrt(magnitude_job))
        else:
            cosine_similarity = 0 # Handle cases with zero magnitude

        job_similarities.append({"resume_id": resume["id"], "similarity": cosine_similarity})
    similarities[jobdescription["id"]] = job_similarities

# You can now print 'similarities' to see the results
# print(similarities)
```
